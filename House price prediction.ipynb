```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#input data
bedrooms = np.array([1,2,3,4]).reshape(-1,1)
prices = np.array([150000,200000,250000,300000])
```

```python
#create a linear regression model
model = LinearRegression()
```

```python
#Train the model
model.fit(bedrooms,prices)
```

```python
#get the coefficients
theta_0 = model.intercept_
theta_1 = model.coef_[0]
```

```python
#make predictions
predictions = model.predict(bedrooms)
```

```python
#plot the data and the best fit line
plt.scatter(bedrooms,prices,label="Actual prices")
plt.plot(bedrooms,predictions,label=f'best fit line:${theta_0:.2f}+${theta_1:.2f})x',color='red')
plt.xlabel('number of bedrooms')
plt.ylabel('House prices')
plt.legend()
plt.show()
```

```python
#use the model to predict the price for a house with 2 bedrooms
newbedrooms = np.array([[2]])
predicted_price = model.predict(newbedrooms)
print(f'predicted price for a house with 2 bedrooms:({predicted_price[0]:.2f}')
```
